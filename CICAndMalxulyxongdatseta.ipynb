```python
import pandas as pd
import glob
import os
import numpy as np

# Specify the parent directory
parent_dir = 'datasets/CICAndMal2017/Benign-CSVs/Benign'

# Get a list of all subdirectories
subdirs = [os.path.join(parent_dir, d) for d in os.listdir(parent_dir) if os.path.isdir(os.path.join(parent_dir, d))]

# List to hold dataframes
df_list = []

for subdir in subdirs:
    # Get a list of all csv files in the subdirectory
    csv_files = glob.glob(os.path.join(subdir, '*.csv'))

    for filename in csv_files:
        df_list.append(pd.read_csv(filename))

# Concatenate all dataframes in the list
benign_df = pd.concat(df_list, ignore_index=True)

# Write the output to a new CSV file
benign_df.to_csv(os.path.join(parent_dir, 'Benign_combined_file.csv'), index=False)
```

```python
import pandas as pd
import glob
import os
import numpy as np

# Specify the parent directory
parent_dirs = ['datasets/CICAndMal2017/Adware-CSVs/Adware', 'datasets/CICAndMal2017/Ransomware-CSVs/Ransomware', 'datasets/CICAndMal2017/Scareware-CSVs/Scareware', 'datasets/CICAndMal2017/SMSmalware-CSVs/SMSmalware']

# List to hold dataframes
df_list = []

for parent_dir in parent_dirs:
    # Get a list of all subdirectories
    subdirs = [os.path.join(parent_dir, d) for d in os.listdir(parent_dir) if os.path.isdir(os.path.join(parent_dir, d))]

    for subdir in subdirs:
        # Get a list of all csv files in the subdirectory
        csv_files = glob.glob(os.path.join(subdir, '*.csv'))

        for filename in csv_files:
            df_list.append(pd.read_csv(filename))

# Concatenate all dataframes in the list
malware_df = pd.concat(df_list, ignore_index=True)

# Write the output to a new CSV file
malware_df.to_csv(os.path.join('datasets/CICAndMal2017', 'Malware_combined_file.csv'), index=False)
```
