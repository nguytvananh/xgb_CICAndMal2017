```python
import numpy as np
import pandas as pd

df = pd.read_csv('datasets/CICAndMal2017/Walnor_combined_file.csv', low_memory=False)
df
```

```python
df[' Label'] = df[' Label'].apply(lambda x: 'MALWARE' if x != 'BENIGN' else x)
df[' Label'].value_counts()
```

```python
df.dtypes[df.dtypes == 'object']
```

```python
df = df.sample(frac=1)
df = df.drop([' Source IP',' Destination IP','Flow ID',' Timestamp'], axis=1)
df
```

```python
df.dtypes[df.dtypes == 'object']
```

```python
df[' Flow IAT Min'] = pd.to_numeric(df[' Flow IAT Min'],errors = 'coerce')
```

```python
df[' Packet Length Std'] = pd.to_numeric(df[' Packet Length Std'],errors = 'coerce')
```

```python
df[' CWE Flag Count'] = pd.to_numeric(df[' CWE Flag Count'],errors = 'coerce')
```

```python
df[' Down/Up Ratio'] = pd.to_numeric(df[' Down/Up Ratio'],errors = 'coerce')
```

```python
df['Fwd Avg Bytes/Bulk'] = pd.to_numeric(df['Fwd Avg Bytes/Bulk'],errors = 'coerce')
```

```python
df.dtypes.value_counts()
```

```python
#Preprocessing
features = df.dtypes[df.dtypes != 'object'].index
df[features] = df[features].apply(
    lambda x: (x - x.mean()) / (x.std()))

#Fill empty values by 0
df = df.fillna(0)
from sklearn.preprocessing import LabelEncoder

X = df.drop([' Label'], axis=1)
y = df.iloc[:, -1]

#encode string class values as integers
label_encoded_y = LabelEncoder().fit_transform(y)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
import matplotlib
matplotlib.use('Agg')
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

model = XGBClassifier()
kfold = KFold(n_splits=10, random_state=None)
results = cross_val_score(model, X, label_encoded_y, cv=kfold)
y_pred = cross_val_predict(model, X, label_encoded_y, cv=kfold)
print(f"Acc : {accuracy_score(label_encoded_y, y_pred)}")
print(f"F1 score : {f1_score(label_encoded_y, y_pred)}")
print(f"Precision score : {precision_score(label_encoded_y, y_pred)}")
print(f"Recall : {recall_score(label_encoded_y, y_pred)}")
```
